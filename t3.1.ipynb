```python
def Calcula_Dist_Freq(lista = [], b = True):
    if b == True:
        absoluta_simples = []
        repetidos = collections.Counter(lista)
        for i in range(len(lista)):
            vfi = repetidos[(lista[i])]
            absoluta_simples.append(vfi)

        total = 0
        for i in range(len(lista)):
            total = total + absoluta_simples[i]
        
        absoluta_acumulada = []
        vaa = 0
        for i in range(len(lista)):
            vaa = vaa + absoluta_simples[i]
            absoluta_acumulada.append(vaa)

        relativa_simples = []
        for i in range(len(lista)):
            vrs = ((absoluta_simples[i])/total)*100
            relativa_simples.append(vrs)

        relativa_acumulada = []
        vra = 0
        for i in range(len(lista)):
            vra = vra + relativa_simples[i]
            relativa_acumulada.append(vra)
            
        # absoluta_acumulada, relativa_simples, relativa_acumulada = Calcula_Freq(lista, absoluta_simples)
        distribuição = pd.DataFrame({ "Mortes": lista,
        "Absoluta simples": absoluta_simples,
         "Relativa Simples": relativa_simples,
         "Absoluta Acumulada": absoluta_acumulada,
         "Relativa Acumulada": relativa_acumulada})

        return distribuição


    if b == False:
        at = lista.max() - lista.min()
        # MÉTODO PRÁTICO
        # 5, se N<25
        # raiz de N, se N>=25
        k = math.sqrt(len(lista))
        h = at/k
        # Arredondamento para MAIS, para não haver vars fora da tabela
        h = math.ceil(h)
        
        intervalos = []
        ocorrencias = []
        menor = round(lista.min(),1)
        menor_amp = round(menor+h,1)
        valor = menor
        while valor < lista.max():
            intervalos.append('{} |- {}'.format(round(valor,1),round(valor+h,1)))
            aux = 0
            for i in range(len(lista)):

                # Tratamento para considerar o ultimo valor
                um = 1
                if (valor+h) == lista.max():
                    um = 0

                # Verifica quantas vars estão dentro daque intervalo    
                if valor <= lista[i] <((valor+h)-um):
                    aux = aux +1

            # Adiciona a quantidade de ocorrencias em cada intervalo        
            if aux != 0:
                ocorrencias.append(aux)
            else:
                ocorrencias.append(0)

            # Salta para o próximo intervalo 
            valor += h

        absoluta_acumulada = []
        vaa = 0
        for i in range(len(ocorrencias)):
            if ocorrencias[i] != 0:
                vaa = vaa + ocorrencias[i]
                absoluta_acumulada.append(vaa)
            else:
                absoluta_acumulada.append(0)

        total = 0
        for i in range(len(ocorrencias)):
            total = total + ocorrencias[i]

        relativa_simples = []
        for i in range(len(ocorrencias)):
            if ocorrencias[i] != 0:
                vrs = ((ocorrencias[i])/total)*100
                relativa_simples.append(vrs)
            else:
                relativa_simples.append(0)

        relativa_acumulada = []
        vra = 0
        for i in range(len(ocorrencias)):
            if relativa_simples[i] != 0:
                vra = vra + relativa_simples[i]
                relativa_acumulada.append(vra)
            else:
                relativa_acumulada.append(0)

        distribuição = pd.DataFrame({    
        "Intervalos": intervalos,
        "Ocorrências": ocorrencias,
        "Relativa Simples": relativa_simples,
        "Absoluta Acumulada": absoluta_acumulada,
        "Relativa Acumulada": relativa_acumulada})

        return distribuição
```
